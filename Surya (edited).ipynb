<h4> About The Spiritualist Newspaper </h4>

*The Spiritualist Newspaper*, first published by E.W. Allen (London) in 1869, forms a key source for how those interested in physical investigations and spiritual forces communicated with the deceased, argued about issues of the day, as well as publicised gatherings. The Newspaper followed a wake of spiritualist activity in Britain, directed by earlier 1850s American Transcendentalism and a more general reaction to scientific naturalism. 

*The Spiritualist* was chosen as an OCR case study for several reasons: 

- Scottish spiritualist histories remain underrepresented as, Foot (2023) discusses. In our case, this partially results from historical detail being contained in heterogeneous and hard-to-transcribe material. This makes accurate OCR recognition more challenging, especially without requisite training data. With a fully searchable sample of 50 pages of *The Spiritualist* enabled through our OCR testing and reference, manually corrected, transcription, spiritualist activity outside of Scotland's Central Belt can be further explored. The trained reference model will also be applied to the entire collection as future work.

- In addition, Foot (2023: 73) suggests that seance mediums underwent 'automatic writing' to channel spiritual forces. The latter insight forms a neat parallel to charting the extent of unexplainable OCR processes in the current technical offering to libraries, as well as the potentially opaque behaviour and outputs of AI-enabled transcription models.

Alongside these two primary reasons, the project also aimed to provide metrics for the standardisation of OCR transcription at the NLS, with the Library currently undergoing a reassessment of its Foxit / ABBYY FineReader licence. The timings registered through the below trials, alongside the CER of outputs, were used as an evidence base for NLS decision-making around digital collection transcription.

 -  **Data format**: png. manuscript images, pre-processed (thresholding and additional cropping) in R Studio using Image Magick. 
 -  **Data creation process**: Vision-Language Model (VLM) Optical Character Recognition.
 - **Data source**: https://data.nls.uk/data/digitised-collections/spiritualist-newspapers/, 50 pages of the first 1869 edition.

<a href = '#section 0'><h4>About the Model</h4></a>

Surya OCR (https://github.com/datalab-to/surya) is the later version of Chandra OCR, developed by Datalab (https://www.datalab.to). It forms a Vision-Language Model (VLM) for OCR processing. Although Datalab cite its performance across over 90 languages, the model is not directly associated with historical document processing and instead is marketed as a business model for document processing. However, Surya's simulatenous processing of image features and text extraction has proven accurate across a wide array of candidates, shown through Hugging Face experimentation (https://huggingface.co/blog/vlms). Datalab also suggest that in using VLM architectures, burdensome - purely linear and sequential - OCR workflows are improved upon.

<a href = '#section 1'><h4>Preparation</h4></a>

**Step 1** - Ensure you have Surya downloaded via the GitHub API, as well as the Tesseract package installed (either through this notebook or your bash terminal), as this forms the model base.

Surya does not have a containerised model, that you can simply import as a neat package, so importation needs to be done separately - with a foundation, recognition and detection predictor each initialised. Surya divides each *Spiritualist* column into bounding boxes and recognises each in turn. This leads to multiple processing lines for each page and can increase operational time, as well as impacting the readability of Jupyter Notebook widgets, used to chart processing progress.

**Step 2** - Other imported libraries seen below enable ALTO XML generation (https://www.loc.gov/standards/alto/), required for hosting *Spiritualist* images in an IIIF-compliant (https://iiif.io) manner via the DHI dashboard, as well as tracking processing time. Pillow (PIL) is also convenient for preprocessing newspaper column structures, to ensure reading order is best preserved, and limiting downstream transcription errors when matched against a reference text.

```python
from PIL import Image
from surya.foundation import FoundationPredictor
from surya.recognition import RecognitionPredictor
from surya.detection import DetectionPredictor

foundation = FoundationPredictor()
recognition = RecognitionPredictor(foundation)
detection = DetectionPredictor()

from pathlib import Path
import xml.etree.ElementTree as ET
import time
```

**Step 3** - With these packages imported, we can upload the *Spiritualist* images and construct a file directory for our OCR transcription outputs.

```python
# Folder setup
image_folder = Path("/Users/joenockels/Documents/Awards/NLS_Fellowship (2025)/Spiritualist_Images") # replace with your own file path
output_folder = Path("./suryaocr_output")
image_folder.mkdir(exist_ok=True)
output_folder.mkdir(exist_ok=True)
```

**Step 4** - Our next stage is to load each of Surya's predictor functions, downloaded from their project GitHub.

```python
# Load Surya models
foundation = FoundationPredictor()
detection = DetectionPredictor()
recognition = RecognitionPredictor(foundation)
```

**Step 5** - Finally, as preparation before OCR processing, define the broad column structure of the material, as well as the ALTO XML scaffold needed for interactive *Spiritualist* image hosting.

```python
# Column splitting function
def split_columns(img_path, columns=3):
    img = Image.open(img_path)
    w, h = img.size
    return [img.crop((i * w // columns, 0, (i + 1) * w // columns, h)) for i in range(columns)]


# ALTO XML generator
def generate_alto_xml(image_width, image_height, predictions, file_name=""):
    ns = "http://www.loc.gov/standards/alto/ns-v4#"
    ET.register_namespace("", ns)

    root = ET.Element("alto", attrib={"xmlns": ns})

    # Description
    desc = ET.SubElement(root, "Description")
    ET.SubElement(desc, "MeasurementUnit").text = "pixel"

    src = ET.SubElement(desc, "SourceImageInformation")
    ET.SubElement(src, "fileName").text = file_name

    # Optional but useful metadata 
    processing = ET.SubElement(desc, "OCRProcessing")
    step = ET.SubElement(processing, "OCRProcessingStep")
    ET.SubElement(step, "processingSoftware").text = "Surya OCR"

    # Layout
    layout = ET.SubElement(root, "Layout")

    page_el = ET.SubElement(layout, "Page", attrib={
        "WIDTH": str(image_width),
        "HEIGHT": str(image_height),
        "PHYSICAL_IMG_NR": "1",
        "ID": "page_1"
    })

    print_space = ET.SubElement(page_el, "PrintSpace", attrib={
        "HPOS": "0",
        "VPOS": "0",
        "WIDTH": str(image_width),
        "HEIGHT": str(image_height)
    })

    # Collect lines with correct coordinates
    all_lines = []

    for page in predictions:
        x_offset = getattr(page, "_x_offset", 0)

        for line in page.text_lines:
            if not hasattr(line, "bbox") or line.bbox is None:
                continue

            x0, y0, x1, y1 = line.bbox

            # Apply column offset
            x0 += x_offset
            x1 += x_offset

            all_lines.append({
                "line": line,
                "bbox": (x0, y0, x1, y1),
                "x_offset": x_offset
            })

    # Preserve reading order by accounting for column offsets in the newspaper layout 
    all_lines.sort(key=lambda l: (l["x_offset"], l["bbox"][1]))

    # One block per column
    blocks = {}

    line_counter = 1
    string_counter = 1

    for item in all_lines:
        line = item["line"]
        x_offset = item["x_offset"]
        x0, y0, x1, y1 = item["bbox"]

        # Create / reuse block per column
        if x_offset not in blocks:
            blocks[x_offset] = ET.SubElement(print_space, "TextBlock", attrib={
                "ID": f"block_{x_offset}",
                "HPOS": str(x_offset),
                "VPOS": "0",
                "WIDTH": str(image_width // 3),
                "HEIGHT": str(image_height)
            })

        text_block = blocks[x_offset]

        text_line = ET.SubElement(text_block, "TextLine", attrib={
            "ID": f"line_{line_counter}",
            "HPOS": str(int(x0)),
            "VPOS": str(int(y0)),
            "WIDTH": str(int(x1 - x0)),
            "HEIGHT": str(int(y1 - y0))
        })

        # Preserve word-level coordinates needed for interactive transcription presentation
        if hasattr(line, "words") and line.words:
            for word in line.words:
                if not hasattr(word, "bbox") or word.bbox is None:
                    continue

                wx0, wy0, wx1, wy1 = word.bbox

                # Apply column offset (this will allow for better presentation and diagnostics)
                wx0 += x_offset
                wx1 += x_offset

                ET.SubElement(text_line, "String", attrib={
                    "ID": f"string_{string_counter}",
                    "CONTENT": word.text,
                    "HPOS": str(int(wx0)),
                    "VPOS": str(int(wy0)),
                    "WIDTH": str(int(wx1 - wx0)),
                    "HEIGHT": str(int(wy1 - wy0))
                })
                string_counter += 1

        else:
            # Fallback (line-level only)
            ET.SubElement(text_line, "String", attrib={
                "ID": f"string_{string_counter}",
                "CONTENT": line.text,
                "HPOS": str(int(x0)),
                "VPOS": str(int(y0)),
                "WIDTH": str(int(x1 - x0)),
                "HEIGHT": str(int(y1 - y0))
            })
            string_counter += 1

        line_counter += 1

    return ET.tostring(root, encoding="utf-8", xml_declaration=True).decode("utf-8")
```

<a href = '#section 2'><h4>OCR Processing</h4></a>

With the input and output folder structure defined, OCR predictor models loaded and functions for splitting *Spiritualist* columns and generating page-level ALTO XML, the following code will put Surya OCR into operation.

**Step 1** - First, we need to upload the required files with a .png extension from our established preparation image folder, as well as start the time package to chart Surya OCR processing. This metric was requested by the NLS Digital Preservation Team, with the Library looking at alternate options to their current Foxit (ABBYYFineReader) licence. For more information on Foxit, as well as current NLS OCR standards and workflows see Gooding et al. (2025, https://doi.org/10.14324/111.9781800088375).

This cell then includes a loop function to process through the *Spiritualist* images, and generate our two exports - the plain .txt file used for calculating CERs, and our ALTO XML.

**Step 2** : Run the OCR, with Surya being a VLM, its predictors will run simulatenously to extract visual image features and textual data, opposed to the more familiar workflow of binarisation, preliminary layout analysis -> line segmentation -> text extraction. This code also stores page level .txt outputs and .xml outputs within our defined folder structure, and calculates each page's processing time, as well as a general 50 page. summary. 

```python
# Gather all PNG images
image_files = sorted(image_folder.glob("*.png"))
print(f"Found {len(image_files)} images.")

start_total = time.time() # begin processing time calculator

for img_path in image_files:
    print(f"\nProcessing {img_path.name}...")
    start_page = time.time()

    combined_text = ""
    combined_predictions =  [] # create an empty list for text outputs

    # Open full image once
    with Image.open(img_path) as img:
        full_width, full_height = img.size
        column_images = split_columns(img_path, columns=3)

    # Process each column, preserving x-offsets for proper ALTO geometry
    for i, col_img in enumerate(column_images):
        x_offset = i * full_width // 3

        preds = recognition([col_img], det_predictor=detection)

        for page in preds:
            # Attach column offset for ALTO coordinates
            page._x_offset = x_offset
            combined_predictions.append(page)

            # Build plain text output
            for line in page.text_lines:
                combined_text += line.text + "\n"

    # Save plain text
    txt_file = output_folder / f"{img_path.stem}.txt"
    txt_file.write_text(combined_text, encoding="utf-8")

    # Generate and save ALTO XML
    alto_xml = generate_alto_xml(
        full_width, full_height, combined_predictions, file_name=img_path.name
    )
    alto_file = output_folder / f"{img_path.stem}.xml"
    alto_file.write_text(alto_xml, encoding="utf-8")

    end_page = time.time()
    print(f"Finished {img_path.name} in {end_page - start_page:.2f} seconds.")

end_total = time.time()
print(f"\nProcessed {len(image_files)} pages in {end_total - start_total:.2f} seconds.")
```

<a href = '#section 3'><h4>Post-Processing</h4></a>

Finally, we concatenate each .txt file into a continous full 50 pg. output. This is needed to comply with our external CER tests using CERberus's local host (https://github.com/WHaverals/CERberus), which provides a more neutral benchmark outside of internal OCR scoring. Developed with the Digital Humanities community in mind, CERberus also accounts for Unicode ranges, variable abbreviations and resulting white space in complex historical manuscripts. We can also export our ALTO XML for interactive text hosting via the DHI.

```python
from pathlib import Path
import shutil

base_folder = Path("/Users/joenockels/nls_ocr_experimentation/nls_ocr_experimentation/notebooks_ocr_experimentation")

alto_source_folder = base_folder / "alto_output"
text_source_folder = base_folder / "suryaocr_output"  # change only if your .txt files are elsewhere

redone_folder = base_folder / "redone_surya"
redone_folder.mkdir(exist_ok=True)

# Copy ALTO XML files into redone_surya
xml_files = sorted(alto_source_folder.glob("*.xml"))

for xml_file in xml_files:
    shutil.copy2(xml_file, redone_folder / f"redone_surya_{xml_file.name}")

# Copy page-level TXT files into redone_surya
txt_files = sorted(
    p for p in text_source_folder.glob("*.txt")
    if not p.name.startswith("redone_surya")
)

for txt_file in txt_files:
    shutil.copy2(txt_file, redone_folder / f"redone_surya_{txt_file.name}")

print(f"Copied {len(xml_files)} ALTO files")
print(f"Copied {len(txt_files)} TXT files")
print(f"Export folder: {redone_folder}")
```
